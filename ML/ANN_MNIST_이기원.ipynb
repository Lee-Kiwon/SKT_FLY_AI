```python
import tensorflow as tf

# 1. MNIST 데이터셋 임포트
fashion_mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# 2. 데이터 전처리
x_train, x_test = x_train/255.0, x_test/255.0

# 3. 모델 구성
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# 4. 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'],)

# 5. 모델 훈련
model.fit(x_train, y_train, epochs=5)

# 6. 정확도 평가
test_loss, test_acc = model.evaluate(x_test, y_test)
print('테스트 정확도:', test_acc)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 데이터 경로 설정
data_dir = '/content/shapes'

# ImageDataGenerator 설정
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# 훈련 데이터셋 생성
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(28, 28),  # 이미지 크기 조정 (원하는 크기로 설정)
    batch_size=32,
    class_mode='categorical',  # 다중 클래스 분류 설정
    subset='training'  # 훈련 데이터에 사용될 부분
)

# 검증 데이터셋 생성
validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(28, 28),
    batch_size=32,
    class_mode='categorical',  # 다중 클래스 분류 설정
    subset='validation'  # 검증 데이터에 사용될 부분
)

# 3. 모델 구성
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28, 3)),
    layers.Dense(28, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(3, activation='softmax')
])

# 4. 모델 컴파일
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 5. 모델 훈련
model.fit(train_generator, epochs=200, validation_data=validation_generator)
```
