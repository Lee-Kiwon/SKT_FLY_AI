```python
import os
import requests
from bs4 import BeautifulSoup

url = "https://n.news.naver.com/mnews/article/052/0001974467?sid=104"
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')

if not os.path.exists("./abc"):
    os.makedirs("./abc")

i = 1

for tag in soup.select('_LAZY_LOADING img'):
    img_url = tag['src']
    save_img = "./abc/" + str(i).zfill(3) + img_url[-8:]
    i += 1

    print(save_img + " : OK")

    headers = {'Referer': url}
    img_data = requests.get(img_url, headers=headers).content

    with open(save_img, 'wb') as f:
        f.write(img_data)
```
